```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import os
from selenium.webdriver.support.select import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
driver = webdriver.Chrome(executable_path=r'C:\Users\LENOVO\Downloads\chromedriver_win32 (1)\Chromedriver.exe')
driver.implicitly_wait(10)
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

# Pages_Links

```python
links=soup.find_all('nav',{'class':'yFHi8N'})
main_Page_link=[]
for i in links:
    main_link=i.find_all('a')
    for j in main_link:
        main_Page_link.append('https://www.flipkart.com'+j.get('href'))
main_Page_link
```

# Model_name

```python
model_name=[]
for i in range(10):
    driver.get(main_Page_link[i])
    m_name=soup.find_all('div',{'class':'col col-7-12'})
    for j in m_name:
        m_name2=j.find_all('div',{'class':'_4rR01T'})
        for k in m_name2:
            model_name.append(k.text)
model_name
```

```python
len(model_name)
```

```python
Model=pd.DataFrame(model_name)
Model
```

```python
Model.to_csv('Model.csv')
```

```python
# Price
```

```python
price=[]
for a in range(10):
    driver.get(main_Page_link[a])
    

    p=soup.find_all('div',{'class':'col col-5-12 nlI3QM'})
    for i in p:
        q=i.find_all('div',{'class':'_3tbKJL'})
        for j in q:
            r=j.find_all('div',{'class':'_25b18c'})
            for k in r:
                s=k.find_all('div',{'class':'_30jeq3 _1_WHN1'})
                for t in s:
                    price.append(t.text)
price
```

```python
len(price)
```

```python
Model_price=pd.DataFrame(price)
Model_price
```

```python
Model_price.to_csv('Model_price.csv')
```

```python
rating=[]

for l in range(10):
    driver.get(main_Page_link[l])
    
    rate=soup.find_all('div',{'class':'col col-7-12'})
    for i in rate:
        s=i.find_all('div',{'class':'gUuXy-'})
        for j in s:
            t=j.find_all('span',{'class':'_1lRcqv'})
            for k in t:
                p=k.find_all('div',{'class':'_3LWZlK'})
                for w in p:
                    rating.append(w.text)
rating
```

```python
Model_rating=pd.DataFrame(rating)
Model_rating
```

```python
Model_rating.to_csv('Model_rating.csv')
```
